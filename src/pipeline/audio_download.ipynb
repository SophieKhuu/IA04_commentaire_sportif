```python
# downloading the audio of the video
import yt_dlp
import os

def download_audio(video_url, output_path):
    ffmpeg_path = r'c:\Users\zebri\Documents\GitHub\IA04_commentaire_sportif\ffmpeg-essentials_build\bin'

    # Créer dossier data s'il n'existe pas
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': output_path,
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'ffmpeg_location': ffmpeg_path  # Specify the path to ffmpeg
    }
    
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([video_url])
```

```python
download_audio('https://www.youtube.com/watch?v=L2LYRXQ4ZCo', 'data/france_argentine.mp3')
```

```python
# transcription with whisper

import whisper

model = whisper.load_model("small")
result = model.transcribe("../../data/france_coree_du_sud.mp3")
print(result["text"])
```
